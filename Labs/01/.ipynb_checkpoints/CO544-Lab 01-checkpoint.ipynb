## Q10. Eigenvalues and Eigenvectors: Au = λu

###  a - the vector contains the eigenvalues 
### M - the array contains the corresponding eigenvectors.

```python
# print(B)

# np.linalg.eig() function to calculate its eigenvalues and eigenvectors
a, M = np.linalg.eig(B)

print("Eigen Values")
print(a)

print("\nEigen Vectors")
print(M)

print("\nTranspose")
print(M.T)

print("\nDot Product of eigen vectors")
# The expression np.dot(M[:,0], M[:,1]) calculates the dot product of the first and second eigenvectors of matrix B
print(np.dot(M[:,0], M[:,1]))

print("\nDot Product of matrices")
# M @ M.T calculates the product of the matrix M with its transpose M.T
print(M @ M.T)
```

## Note how the choice of the number of bins and the number of samples affects the histogram you observe.
Here is a bit of code to help you study this.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the experiment
MaxTrials = 10
NumSamples = 200
NumBins = 20

# Loop over a set number of trials
for trial in range(MaxTrials):
    # Generate a set of random data samples
    x = np.random.rand(NumSamples, 1)
    
    # Create a new figure for each trial
    plt.figure(figsize=(10, 5))
    
    # Generate a histogram of the data using the specified number of bins
    counts, bins, patches = plt.hist(x, bins=NumBins, color='b', alpha=0.8, rwidth=0.8)
    
    # Save the current figure with a unique filename for each trial
    plt.savefig('fig-2.2-'+str(trial)+'.png')
    
    # Calculate the variation within bin counts and print the result
    print("Variation within bin counts: ", np.var(counts/NumSamples))
```

## Repeat the above with 1000 random numbers drawn from a Gaussian distribution of mean 0 and standard deviation 1 using x = random,randn(1000, 1);.

Now try the following where we add and subtract uniform random numbers:

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

# create an array of zeros with length N
x1 = np.zeros(N)

# fill the array x1 with random values
for n in range(N):
  x1[n] = sum(np.random.rand(120000, 1)) - np.sum(np.random.rand(120000, 1))

# create a histogram of x1 with 40 bins
plt.hist(x1, 40, color='b', alpha=0.8, rwidth=0.8)

plt.xlabel('Bin', fontsize=16)
plt.ylabel('Counts', fontsize=16)
plt.grid(True)
plt.title('Histogram', fontsize=16)
plt.show()
```

# Part-3. 
# Uncertainty in Estimation

Much of what we do in Machine Learning is estimating parameters of a model from a finite set of data.
Here is why one might think having more data is a good thing: Consider estimating the variance of a
uni-variate Gaussian distribution from samples drawn from it. If we did this from different relations of
the samplings, we would expect slightly different answers (every time we do it). We would, of course,
like this variation to be small. Figure 1 shows how the variance of the estimation varies with sample
size. Here is a snippet of code to study this.

```python
sampleSizeRange = np.linspace(100, 200, 40)
plotVar = np.zeros(len(sampleSizeRange))

for sSize in range(len(sampleSizeRange)):
    numSamples = int(sampleSizeRange[sSize])
    
    MaxTrial=2000
    vStrial=np.zeros(MaxTrial)
    
    for trial in range(MaxTrial):
        xx = np.random.randn(numSamples,1)
        vStrial[trial] = np.var(xx)
        
    plotVar[sSize] = np.var(vStrial)
    
plt.plot(sampleSizeRange, plotVar)
```

## 4. Bivariate Gaussian Distribution

Here are two functions to help us calculate and plot contours on a multivariate Gaussian density.

```python
def gauss2D(x, m, C):
    Ci = np.linalg.inv(C)
    dC = np.linalg.det(Ci)
    num = np.exp(-0.5 * np.dot((x-m).T, np.dot(Ci, (x-m))))
    den = 2 * np.pi * dC
    return num/den

def twoDGaussianPlot (nx, ny, m, C):
    x = np.linspace(-5, 5, nx)
    y = np.linspace(-5, 5, ny)
    X, Y = np.meshgrid(x, y, indexing='ij')
    
    Z = np.zeros([nx, ny])
    
    for i in range(nx):
        for j in range(ny):
            xvec = np.array([X[i,j], Y[i,j]])
            Z[i,j] = gauss2D(xvec, m, C)
            
    return X, Y, Z
```

```python
nx, ny = 50, 40

plt.figure(figsize=(6,6))

m1 = np.array([0, 2])
C1 = np.array([[2, 1], [1,2]], np.float32)
Xp, Yp, Zp = twoDGaussianPlot(nx, ny, m1, C1)

plt.contour(Xp, Yp, Zp, 3)
plt.grid(True)
```

# 5. Sampling from a Multivariate Gaussian Distribution

Consider the covariance matrix C =[[2 1][1 2]]
Factorize this into a lower trinagular matrix and its transpose (Cholosky decomposition), At A = C,
and confirm the factorization is correct by multiplying.

```python
C = [[2, 1], [1, 2]]
print(C)

# computes the Cholesky decomposition of a positive definite matrix C.
# A is a lower triangular matrix, such that C = A @ A.T (A*A.T)
A = np.linalg.cholesky(C)

print(A) # lower triangular matrix
print(A @ A.T) # must be equal to C
```

Generate 10000 bivariate Gaussian random numbers by X = np.random.randn(1000, 2);
Transform each of the two dimensional vectors (rows of X) by Y = X @ A.

```python
X = np.random.randn(10000, 2)
Y = X @ A
print(X.shape)
print(Y.shape)
```

Now draw a scatter plot of X and Y.

```python
plt.scatter(Y[:, 0], Y[:, 1], s=3, c='m')
plt.scatter(X[:, 0], X[:, 1], s=3, c='c')
plt.grid(True)
plt.title("Scatter of Isotropic and Correlated Gaussian Densities")
```

# 6. Distribution of Projections

Construct a vector u = [sin θ cos θ], parameterized by the variable θ.

```python
theta = np.pi / 3
u = [np.sin(theta), np.cos(theta)]

print("The vector: ", u)
print("Sum of squares: ", u[0]**2 + u[1]**2)
print("Degrees: ", theta*180 / np.pi)
```

Compute the variance of projections of the data in X along this direction:

```python
yp = Y @ u
print(yp.shape)
print("Projected Variance: ", np.var(yp))
```

Plot how this projected variance changes as a function of θ:

```python
nPoints = 50;
pVars = np.zeros(nPoints)
thRange = np.linspace(0, 2*np.pi, nPoints)

for n in range(nPoints):
  theta = thRange[n]  
  u = [np.sin(theta), np.cos(theta)]
  pVars[n] = np.var( Y @ u )
  
plt.plot(pVars)
plt.grid(True)
plt.xlabel("Direction", fontsize=14)
plt.ylabel("Variance of Projections", fontsize=14)
```
